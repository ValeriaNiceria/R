```r
# Criando um modelo
modelo <- lm(speed ~ dist, data = cars)
modelo
```

```r
plot(modelo)
```

```r
plot(speed ~ dist, data = cars)
```

```r
plot(speed ~ dist, data = cars)
abline(modelo)
```

## Como prever?
### previsão = intersecção + (inclinação * valor a prever)

```r
modelo$coefficients
```

```r
intercept <- modelo$coefficients[1]
dist <- modelo$coefficients[2]

previsao = intercept + dist * 22
previsao
```

```r
previsao = predict(modelo, data.frame(dist=22))
previsao
```

```r
# Sumário do modelo
summary(modelo)
```

### Residuais => É a distância entre o ponto dos dados e a linha de melhor ajuste

```r
modelo$residuals
```

```r
# Valores ajustados
modelo$fitted.values
```

```r
plot(modelo$fitted.values, cars$dist)
```
